```python
import trimesh
import numpy as np

import matplotlib.pyplot as plt

# Load the 3D model
mesh = trimesh.load_mesh('bow.obj')

# Extract vertices and faces
vertices = mesh.vertices
faces = mesh.faces

# Plot the 3D model using matplotlib
fig = plt.figure()
fig.set_size_inches(30,30)
ax = fig.add_subplot(projection='3d', box_aspect=(8, 60, 80))

# Plot the mesh
ax.plot_trisurf(vertices[:, 0], vertices[:, 1], vertices[:, 2], triangles=faces)


# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
